Analyzing COVID-19 Data

The Coronavirus disease orgininate in Wuham, China in 2019. This virus swept across the world and is now classifie as a Pandemic.The World health Organization has implimented many counter measures like social distancing and ost of the world is in "quarantine".

```python
%matplotlib inline
import pandas
```

Imported Data from S3 Bucket

```python
df = pandas.read_excel('s3://nassiri-ia241/COVID-19-geographic-disbtribution-worldwide-2020-04-28.xlsx')
df
```

The imported COVID-19 data, all 13,623 cases

The Sum of Death Cases Per Day

```python
sum_death_by_date = df.groupby('dateRep').sum()['cases']

sum_death_by_date.plot()
```

A graph that represents death cases per day, this graph shows the number of deaths per day is at its highest in the month of April.

The Top 10 Countries with the highest Deaths

```python
sum_death_by_country = df.groupby('countriesAndTerritories').sum()['deaths']
sum_death_by_country.nlargest(10).plot.bar()
```

A graph that represents the top 10 countries with the highest deaths. As seen in the graph that United States has the highest amount of deaths & Brazil has the lowest amount.

The List of the Unique Countries and Territories

```python
pandas.unique(df['countriesAndTerritories'])
```

This is an array of uniques countries and territories that was imported from pandas.

Total Cases in China (Chinas Data)

```python
chn_data = df.loc[df['countriesAndTerritories'] == 'China']
chn_data
```

This is part of the data that represents only Chinas data. The majority of death cases in China are in the months of January and April.

Number of Cases that are Related to the Number of Deaths

```python
chn_data.plot.scatter(x='cases',y='deaths')
```

This is Scatter plot using only Chinas data to compare the number of cases to the number of deaths to find a relation.
